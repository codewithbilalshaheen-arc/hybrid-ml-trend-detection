This notebook:
1. Loads the time series data from Student A
2. Visualizes the top 5 topics over time
3. Assigns human-readable labels to all topics

## 1. Setup & Load Data

```python
import pandas as pd
import plotly.graph_objects as go

# Load the time series data
topic_ts = pd.read_csv('topic_time_series.csv')
topic_ts['week'] = pd.to_datetime(topic_ts['week'])

print('✅ Data loaded!')
print(f'Shape: {topic_ts.shape}')
print(f'Date range: {topic_ts[\"week\"].min()} to {topic_ts[\"week\"].max()}')
print(topic_ts.head())
```

## 2. Topic Labels

```python
# Load topic labels
topic_labels = pd.read_csv('topic_labels.csv')

print('📋 All Topics:')
print('='*80)
for _, row in topic_labels.iterrows():
    topic_id = row['topic_id']
    label = row['label']
    top_words = row['top_words'].split(', ')[:8]
    print(f'Topic {topic_id:2d}: {label}')
    print(f'  Words: {\" | \".join(top_words)}')
```

## 3. Top 5 Topics Analysis

```python
# Calculate total posts per topic
topic_totals = topic_ts.iloc[:, 1:].sum().sort_values(ascending=False)

print('📊 Total Posts by Topic (Ranked):')
print('='*50)
for idx, (topic, total) in enumerate(topic_totals.items(), 1):
    topic_num = int(topic.split('_')[1])
    label = topic_labels[topic_labels['topic_id'] == topic_num]['label'].values[0]
    print(f'{idx}. Topic {topic_num:2d}: {int(total):6d} posts - {label}')
```

## 4. Visualize Top 5 Topics

```python
# Get top 5
top_5_topics = topic_totals.head(5).index.tolist()
top_5_nums = [int(t.split('_')[1]) for t in top_5_topics]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

# Create charts
for idx, topic_id in enumerate(top_5_nums):
    topic_label = topic_labels[topic_labels['topic_id'] == topic_id]['label'].values[0]
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=topic_ts['week'],
        y=topic_ts[f'topic_{topic_id}'],
        mode='lines+markers',
        name=topic_label,
        line=dict(color=colors[idx], width=3),
        marker=dict(size=6),
        fill='tozeroy'
    ))
    
    fig.update_layout(
        title=f'<b>Topic {topic_id}: {topic_label}</b>',
        xaxis_title='Week',
        yaxis_title='Number of Posts',
        template='plotly_white',
        height=450
    )
    
    fig.show()
```
